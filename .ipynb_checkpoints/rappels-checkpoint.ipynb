```python
dico_list = []

# Iterate trough pages

for year in range(2009, 2019):
    print(year)

    for page_number in range(1, 11):
        print('PAGE NUMBER ', page_number+1 )

        url = "https://www.konsum.admin.ch/bfk/fr/home/produktesicherheit/produkterueckrufe/archiv1/produktrueckrufe-und-sicherheitsinformationen-"+str(year)+"/_jcr_content/nsbnewslist.entries.html?pageIndex="+str(page_number)+"&startDate=01.01."+str(year)+"&endDate=31.12."+str(year)+"&organization=1008%2C1079&topic="

        print(url)
        try:
            html = requests.get(url, headers=request_headers)
            soup = BeautifulSoup(html.content, "lxml")



            # Pour chaque rappel
            for item in soup.find_all('div', {'class': 'list-group-item'}):

                for lien in item.find_all('a'):
                    print(lien.get_text(strip=True))

                    whole_link = "https://www.konsum.admin.ch/" + lien['href']

                    print(whole_link)

                    dico = {}
                    dico['titel'] = lien.get_text(strip=True)
                    dico['url'] = whole_link
                    dico['page_number'] = page_number + 1
                    dico['year'] = year

                    # Create dictionary

                    dico_list.append(dico)

        # Deal with unexisting pages    
        except Exception as e:
            print(e)
            continue
        
 
```
