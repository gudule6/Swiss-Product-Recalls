```python
dico_list = []

# Iterate trough pages

for year in range(2009, 2019):
    print(year)

    for page_number in range(0, 11):
        print('PAGE NUMBER ', page_number+1 )

        
        url = "https://www.konsum.admin.ch/bfk/fr/home/produktesicherheit/produkterueckrufe/archiv/produktrueckrufe-und-sicherheitsinformationen-"+str(year)+"/_jcr_content/nsbnewslist.entries.html?pageIndex="+str(page_number)

        print(url)
        try:
            html = requests.get(url, headers=request_headers)
            soup = BeautifulSoup(html.content, "lxml")



            # Pour chaque rappel
            for item in soup.find_all('div', {'class': 'list-group-item'}):

                for lien in item.find_all('a'):
                    print(lien.get_text(strip=True))

                    whole_link = "https://www.konsum.admin.ch/" + lien['href']

                    print(whole_link)

                    dico = {}
                    dico['titel'] = lien.get_text(strip=True)
                    dico['url'] = whole_link
                    dico['page_number'] = page_number + 1
                    dico['year'] = year

                    # Create dictionary

                    dico_list.append(dico)

        # Deal with unexisting pages    
        except Exception as e:
            print(e)
            continue
        
print(len(dico_list))
```

```python
doc = pd.Series(' '.join(df['titel']).lower().split())
```

```python
type(doc)
```

```python
# Créer une liste
doc = [word for word in doc]
# Créer une string
doc = ' '.join(doc)
```

```python
# Spacy magic
pos_counts = defaultdict(Counter)
doc = nlp(doc)
```

```python
for token in doc:
    print(token, token.lemma_)
```

```python
lems = [word.lemma_ for word in doc if word.is_stop != True and word.is_punct != True]
```

```python
word_freq = Counter(lems)
common_words = word_freq.most_common(100)
```
